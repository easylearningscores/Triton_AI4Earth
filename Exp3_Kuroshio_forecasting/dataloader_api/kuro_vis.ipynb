```python
import matplotlib.pyplot as plt
import numpy as np
import netCDF4 as nc
from matplotlib.animation import FuncAnimation

data = nc.Dataset('/jizhicfs/easyluwu/ocean_project/kuro/KURO.nc')
vgos = data['vgos']
ugos = data['ugos']
print(vgos.shape, ugos.shape)

time_step = 500  

data_slice = ugos[time_step, :, :]

plt.figure(figsize=(10, 8))

im = plt.imshow(data_slice, origin='lower', cmap='RdBu_r')  
plt.colorbar(im, label='m/s')
plt.title(f"Surface Geostrophic Velocity (ugos) - Time Step {time_step}")

plt.show()
```
